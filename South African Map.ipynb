## Using folium to Map South African Provinces

```python
import folium
```

```python
SouthAfrica_map = folium.Map(location =[-30.5595,22.9375],zoom_start = 4)
```

```python
WesternCape = folium.map.FeatureGroup()
WesternCape.add_child(folium.features.CircleMarker([-33.2278,+21.8569],radius = 2, color = 'red', fill_color = 'red'))
SouthAfrica_map.add_child(WesternCape)

Gauteng = folium.map.FeatureGroup()
Gauteng.add_child(folium.features.CircleMarker([-26.2708,28.1123],radius = 2,color = 'red', fill_color = 'red'))
SouthAfrica_map.add_child(Gauteng)

KZN = folium.map.FeatureGroup()
KZN.add_child(folium.features.CircleMarker([-28.5306,30.8958],radius = 2,color = 'red', fill_color = 'red'))
SouthAfrica_map.add_child(KZN)

NorthWest = folium.map.FeatureGroup()
NorthWest.add_child(folium.features.CircleMarker([-26.6639,25.2838],radius = 2,color = 'red', fill_color = 'red') )
SouthAfrica_map.add_child(NorthWest)

Limpopo = folium.map.FeatureGroup()
Limpopo.add_child(folium.features.CircleMarker([-23.4013, 29.4179],radius = 2,color = 'red', fill_color = 'red') )
SouthAfrica_map.add_child(Limpopo)

FreeState = folium.map.FeatureGroup()
FreeState.add_child(folium.features.CircleMarker([-28.4541,26.7968],radius = 2,color = 'red', fill_color = 'red') )
SouthAfrica_map.add_child(FreeState)

NorthernCape = folium.map.FeatureGroup()
NorthernCape.add_child(folium.features.CircleMarker([-29.0467, 21.8569],radius = 2,color = 'red', fill_color = 'red') )
SouthAfrica_map.add_child(NorthernCape)

EasternCape = folium.map.FeatureGroup()
EasternCape.add_child(folium.features.CircleMarker([-32.2968,26.4194],radius = 2,color = 'red', fill_color = 'red') )
SouthAfrica_map.add_child(EasternCape)

Mpumalanga = folium.map.FeatureGroup()
Mpumalanga.add_child(folium.features.CircleMarker([-25.5653,30.5279],radius = 2,color = 'red',fill_color = 'red'))
SouthAfrica_map.add_child(Mpumalanga)
```
